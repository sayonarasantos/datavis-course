#### Dados utilisados (Coleta aleatória)
Para realizar a coleta a aleatória de dados exigida na questão, executou-se um sorteio de 10 programas de cada unidade federativa (UF).

```python
from random import randint

# Redução do banco para apresentar apenas as unidades federativas e a quantidade de bolsas
ds2 = dataset[['UF','Total Linha']]

# Escolha aleatória dos dados e obtenção do total de bolsas em cada 10 projetos por UF
ufs = ['AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA',
       'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
lista = [] # lista com todos projetos sorteados [UF, Total Linha]
listaSoma = [] # lista com o total de bolsas por UF em cada 10 projetos

for uf in ufs:
    dsUF = ds2[ds2['UF']==uf].reset_index() # conjunto de infos de cada UF
    for i in range(10):
        sorteio = randint(0,len(dsUF.index))
        lista.append(dsUF.iloc[sorteio])
print lista[1:]
#     print pd.DataFrame(lista)
#     listaTotal.append(sum(lista))
```

```python
from random import randint

# Redução do banco para apresentar apenas as unidades federativas e a quantidade de bolsas
ds2 = dataset[['UF','Total Linha']]

# Escolha aleatória dos dados e obtenção do total de bolsas em cada 10 projetos por UF
ufs = ['AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA',
       'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
lista = [] # lista com o total de bolsa de 10 projetos de cada uf

for uf in ufs:
    dsUF = ds2[ds2['UF']==uf].reset_index()
    soma = 0 # soma do total de bolsas dos 10 projetos de uma uf
    for i in range(10):
        sorteio = randint(0,len(dsUF.index))
        soma += dsUF['Total Linha'].iloc[sorteio]
    lista.append([uf,soma])
# lista
```

```python
dataset4 = lista[['UF','Total Linha']]
listaUF = []
listaTotal = []
for uf in ufs:
    ds = dataset4[dataset4['UF']==uf]
    listaTotal.append(ds['Total Linha'].sum())
    listaUF.append(uf)
dataset5 = pd.DataFrame([listaTotal, listaUF], index=['Total Linha','UF'])

```

```python
dataset4 = lista[['UF','Total Linha']]
listaUF = []
listaTotal = []
for uf in ufs:
    ds = dataset4[dataset4['UF']==uf]
    listaTotal.append(ds['Total Linha'].sum())
    listaUF.append(uf)
dataset5 = pd.DataFrame([listaTotal, listaUF], index=['Total Linha','UF'])
```
