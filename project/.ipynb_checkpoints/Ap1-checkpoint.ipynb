### Coleta aleatória
Para realizar a coleta a aleatória de dados exigida neste trabalho, executou-se um sorteio de 10 programas de cada unidade federativa (UF).

```python
from random import randint

# Redução do conjunto para apresentar apenas as unidades federativas e a quantidade de bolsas
ds2 = dataset[['UF','Total Linha']]

# Escolha aleatória dos dados e obtenção do total de bolsas em cada 10 projetos por UF
ufs = ['AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA',
       'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
listaSoma = [] # lista com o total de bolsa de 10 projetos de cada uf
listaUF = [] # lista das ufs que receberam bolsas

for uf in ufs:
    dsUF = ds2[ds2['UF']==uf].reset_index()
    soma = 0 # soma do total de bolsas dos 10 projetos de uma uf
    for i in range(10):
        sorteio = randint(0,len(dsUF.index))
        soma += dsUF['Total Linha'].iloc[sorteio]
    listaSoma.append(soma)
    listaUF.append(uf)
# print listaSoma
# print listaUF
```
