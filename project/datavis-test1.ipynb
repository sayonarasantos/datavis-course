```python
dataset1 = pd.read_csv('capes_bolsas.csv')
dataset1
```

```python
dataset1 = dataset1.fillna(0)
```

```python
dataset1.loc[1,:]
```

```python
# Redução do banco para apresentar apenas as unidades federativas e a quantidade de bolsas
ds1_2 = dataset1[['UF','Total Linha']]

ufs = ['AL', 'AM', 'AP', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MG', 'MS', 'MT', 'PA',
       'PB', 'PE', 'PI', 'PR', 'RJ', 'RN', 'RO', 'RR', 'RS', 'SC', 'SE', 'SP', 'TO']
listOfSums = []

for uf in ufs:
    columnSum = 0
    dsUF = ds1_2[ds1_2['UF']==uf].reset_index() # conjunto de infos de cada UF
    columnSum = dsUF['Total Linha'].sum()
    listOfSums.append(columnSum)
ds_SumByUF = pd.DataFrame([ufs, listOfSums], index=['UF','Total Linha'])
ds_SumByUF
```

```python
ds_SumByUF.to_csv('bolsasPorEstado.csv', sep=',', encoding='utf-8')
```

```python
listOfSums3 = []
listOfSums32 = []
listOfSums33 = []
for index, row in ds3_1.iterrows():
    totalY = 0
    totalY = row['GRADUAÇÃO']+row['GRADUAÇÃO PLENA']+row['GRADUAÇÃO SANDUÍCHE']+row['Graduação Sanduíche']
    listOfSums3.append(totalY)
    listOfSums32.append('grad')
    listOfSums33.append(row['Ano'])

# ds3_1['Ano'].drop_duplicates()
ds_Year_grad = pd.DataFrame([listOfSums33, listOfSums32, listOfSums3], index=['Ano', 'Tipo','Total Linha'])
ds_Year_grad
```
